```
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

import io 

  
df = pd.read_csv(io.BytesIO(uploaded['Data_Handling.csv'])) 
print(df)
```

1.This is to show the dependency of season and its factors effect on the number of total number of people (Casual + registered) who use the cycles Using a bar charts


```
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.barplot(x = 'season', y = 'cnt',ci = None, data = df)

plt.savefig('Season_vs_cnt.png')
```

```
sns.lineplot(x = 'temp', y = 'cnt', data = df)
```

```
sns.set(rc={'figure.figsize':(11,8)})
sns.lineplot(x = 'hum', y = 'cnt',ci = None, data = df)
```

```
sns.lineplot(x = 'windspeed', y = 'cnt', data = df)
```

2.Peak hours of sales during the day and difference in demand for bikes during holidays vs working days


```
sns.set(rc={'figure.figsize':(15,8)})
sns.boxplot(x = 'hr', y='cnt', data = df)
```

```
sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(x = 'hr', y='cnt',hue = 'workingday', ci = None, data = df)
```

```
sns.barplot(x = 'hr', y='cnt',hue = 'holiday', ci = None, data = df)
```
